```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

pd.set_option('display.max_rows', None)

file_path='D:/Calculus Research/Relatórios Setoriais/Estatísticas Setoriais/Consolidado SFN.xlsx'
```

##### * CARTEIRA DE CREDITO -------------

```python
##### * CARTEIRA DE CREDITO -------------

BaseSaldoTotal = pd.read_excel(file_path, sheet_name='1.0BaseSaldoTotal', parse_dates=True, index_col=None)
print(BaseSaldoTotal.columns)
BaseSaldoTotal.info()
BaseSaldoTotal.columns=['Data','SaldoTotal','SaldoPJTotal','SaldoPFTotal','SaldoRLivres','SaldoRDTotal']

#Filtro de Data
BaseSaldoTotal_Graph = BaseSaldoTotal[BaseSaldoTotal['Data'].dt.year >= 2010]

plt.figure(figsize=(10, 6))
plt.stackplot(BaseSaldoTotal_Graph['Data'], BaseSaldoTotal_Graph['SaldoPJTotal']/1000000, BaseSaldoTotal_Graph['SaldoPFTotal']/1000000,
labels=['Pessoa Jurídica', 'Pessoa Física'],colors=['blue', 'red'])

plt.title('Carteira de Crédito do SFN', fontsize=14)
plt.ylabel('R$tri', fontsize=12)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2, fontsize=12)

# Configurando o eixo x para exibir apenas os anos com espaçamento de 5 anos
plt.gca().xaxis.set_major_locator(mdates.YearLocator(3))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=0, fontsize=12)  # Rotação e tamanho da fonte para os ticks do eixo x
plt.yticks(rotation=0, fontsize=12)  # Rotação e tamanho da fonte para os ticks do eixo y
plt.tight_layout()

plt.savefig('D:/Calculus Research/Traders Club/Relatórios/Bancos/Macro_Cart_Credito_py.jpeg', format='jpeg', dpi=300)

plt.show()
```
